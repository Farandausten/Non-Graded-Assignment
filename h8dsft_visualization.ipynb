```python
df = pd.read_csv('data.csv')
```

# BARPLOT

```python
Pemainhebat = df[(df["Composure"] >= 70)&
   (df["Marking"] >= 40) & 
   (df["StandingTackle"] >= 40) &
   (df["SlidingTackle"] >= 40)&
   (df["Finishing"] >= 80) & 
   (df["Crossing"] >= 75)]
```

```python
df_banding = df[df['Name'].isin(['L. Messi', 'Cristiano Ronaldo'])]['Composure']
```

```python
Pemainhebat['Composure'].plot(kind='bar', figsize=(10, 10))
 # add to x-label to the plot
plt.ylabel('COMPOSURE') # add y-label to the plot
plt.title('DUA PEMAIN JAGO') # add title to the plot
plt.xlabel('Pemain')
plt.show()
```

# HISTOGRAM

```python
count, bin_edges = np.histogram(Pemainhebat['Composure'])
```

```python
Pemainhebat['Composure'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)
```

```python
pd.set_option("display.max_columns", None)
```

# PIE CHART

HARUS PAKE AGG

```python
Pemainhebat['Composure'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° (Africa)
                            shadow=True,       # add shadow      
                            )

plt.title('Perbandingan Composure')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

# SCATTER PLOT

```python
Pemainhebat.plot(kind='scatter', x = 'Name', y = 'Overall' , figsize=(10, 6), color='darkblue')

plt.title('Total Orang Cupu')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

# BOX PLOT

```python
Pemainhebat['Composure'].plot(kind='box', figsize=(8, 6))

plt.title('Box plot Pemain Hebat')
plt.ylabel('Number')

plt.show()
```
