# BARPLOT

```python
Pemainhebat = df[(df["Composure"] >= 70)&
   (df["Marking"] >= 40) & 
   (df["StandingTackle"] >= 40) &
   (df["SlidingTackle"] >= 40)&
   (df["Finishing"] >= 80) & 
   (df["Crossing"] >= 75)]
```

```python
ggwp = df.loc[ : 12, ['Name', 'Age', 'Overall', 'Composure', 'Finishing', 'Nationality', 'Club']]
ggwp1 = df.loc[ : 50, ['Name', 'Age', 'Overall', 'Composure', 'Finishing', 'Nationality']]
```

```python
plt.figure(figsize=(8, 6))
sns.factorplot(x = 'Name', y = 'Overall', kind = 'bar', data = ggwp)
plt.xlabel("Day")
plt.title("Total Bill Per Day of Week")
plt.show()
```

# HISTOGRAM, KDE, DENSITY

```python
plt.figure(figsize=(8, 6))
sns.histplot(x = 'Name', y = 'Overall', data = ggwp)
plt.xlabel("Day")
plt.title("Total Bill Per Day of Week")
plt.show()
```

```python
sns.distplot(ggwp['Composure'])
sns.distplot(ggwp['Overall']);
```

```python
with sns.axes_style('white'):
    sns.jointplot("Composure", "Overall", ggwp, kind='kde');
```

```python
with sns.axes_style('white'):
    sns.jointplot("Composure", "Overall", ggwp, kind='hex');
```

# PIE CHART

HARUS PAKE AGG

```python
plt.figure(figsize=(8, 6))
sns.set_style('whitegrid')
sns.swarmplot(x='Overall', y='Name', data=ggwp1)
plt.xlabel("Size")
plt.ylabel("Total Bill")
plt.title("Total bill per size of the table")
plt.show()
```

# PAIR PLOT

```python
sns.pairplot(ggwp, hue='Nationality', size=2.5);
```

# VIOLIN PLOT

```python
plt.figure(figsize=(10, 7))
sns.violinplot(x='Name', y='Composure', hue="Club",
              data=ggwp, palette="muted")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Total Bill per Day of the Week")
plt.show()
```
